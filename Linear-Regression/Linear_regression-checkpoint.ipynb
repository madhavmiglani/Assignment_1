```python
import numpy as np
import matplotlib.pyplot as plt

#Given values
w=1 
b=0
mean = 0
variance =1

# Dataset
x=[]
y=[]

for i in range(1000):
    x.append(np.random.random()*10)
    y.append(w*x[-1] + b + np.random.normal(0,1))

```

```python
# Displaying the plot
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#y_i = Predicted values
y_i = []

# y_i = a + b*x
a = np.random.random()
b = np.random.random()

# Compute derivative of cost function w.r.t b and set it equal to 0
def update_b(a,b,x,y):
    h = np.sum(np.multiply(x,x))
    k = np.sum(np.multiply(a,x)- np.multiply(y,x))
    b_new = -1 * k / h
    return b_new

# Compute derivative of cost function w.r.t a and set it equal to 0
def update_a(a,b,x,y):
    h = np.sum(np.multiply(b,x)-y)
    n=1000
    a_new = -1*h/n
    return a_new
```

```python
# Plot best fit line
def plot_best_fit_line(x,y,y_i,a,b):
    y_i = np.multiply(b,x) + a   
    plt.plot(x, y_i, color ='red', label = 'Best Fit Line')
    plt.scatter(x,y, color ='yellow')
    plt.legend()
    plt.show()

a_new = update_a(a,b,x,y)
b_new = update_b(a,b,x,y)

# Ensuring that the values of a and b are such that they satisfy both derivatives of w.r.t a and b to 0
if a_new == a and b_new ==b:
    a= a_new
    b= b_new
    plot_best_fit_line(x,y,y_i,a,b)

else:
    while a_new !=a or b_new !=b:
        if a_new !=a:
            a=a_new
            b_new = update_b(a,b,x,y)
            if(b_new == b):
                plot_best_fit_line(x,y,y_i,a,b)
            else:
                b=b_new
                a_new=update_a(a,b,x,y)
                b_new = update_b(a,b,x,y)
        else:
            b=b_new
            a_new = update_a(a,b,x,y)
            if(a_new == a ):
                plot_best_fit_line(x,y,y_i,a,b)
            else:
                a=a_new
                a_new= update_a(a,b,x,y)
                b_new = update_b(a,b,x,y)

```
